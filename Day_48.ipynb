```python
#Coding Exercise: Generative AI Applications (Text Summarization and Content Generation)
#Write a Python program using the Hugging Face 'transformers' library to summarize a short paragraph. Use any open source LLM model from the hub.

from transformers import pipeline

summarizer = pipeline("summarization")

text = """
The rise of artificial intelligence (AI) has led to significant advancements in various industries, such as healthcare, finance, and transportation.
AI is being used to automate tasks, analyze large datasets, and assist in decision-making. While there are many benefits, there are also concerns about the potential impact of AI on jobs, privacy, and ethics.
It is essential for policymakers, businesses, and individuals to navigate these challenges to ensure the responsible development and deployment of AI technologies.
"""

summary = summarizer(text, max_length=50, min_length=25, do_sample=False)

print("Original Text:\n", text)
print("\nSummarized Text:\n", summary[0]['summary_text'])

```
